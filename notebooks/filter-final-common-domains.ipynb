```python
import pandas as pd
import numpy as np
```

```python
final_common_domains = pd.read_csv("final_common_domains.csv")
```

```python
final_common_domains.head()
```

```python
countries = np.unique(np.hstack((final_common_domains['Country 1'].unique(), final_common_domains['Country 2'].unique())))
countries
```

```python
for c in countries:
    curr_df = final_common_domains.loc[(final_common_domains['Country 1'] == c) | (final_common_domains['Country 2'] == c)]
    curr_df.drop(columns = ['Unnamed: 0'], inplace = True)
    curr_df.to_csv('common_domains/' + c + '-common-domains.csv')
```
